##Matplotlib<br>
&nbsp;&nbsp;&nbsp;Matplotlib is a Python 2D plotting library which produces
publication-quality figures in a variety of hardcopy formats
and interactive environments across
platforms.

**Use the following convention**

```python
import matplotlib.pyplot as plt
```

**Import numpy to get data for our plots**

```python
import numpy as np
```

```python
# plotting the data
x =np.linspace(0,10,100)
y =np.sin(x)
z =np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

##Plotting the data

**Sine wave**

```python
plt.plot(x,y)
plt.show()
```

**Cosine wave**

```python
plt.plot(x,z)
plt.show()
```

##Adding title , x- axis & y-axis labels


**Sine wave**

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine wave')
plt.show()
```

**Cosine wave**

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('Cosine wave')
plt.show()
```

##Parabola

```python
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y,'r+') # red ,+ symbol
plt.show()
```

```python
plt.plot(x,y,'bx') # blue , x symbol
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g--')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

##Bar Plot

```python
fig  = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages =['English','French','Spanish','Latin','German']
people = [100,50,150,50,40]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of people')
plt.show()
```

##Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages =['English','French','Spanish','Latin','German']
people = [100,50,150,50,40]
ax.pie(people ,labels =languages,autopct='%1.1f%%')
plt.show()
```

##Scatter Plot

```python
x =np.linspace(0,10,30)
y =np.sin(x)
z = np.cos(x)
```

```python
fig2 = plt.figure()
ax =fig2.add_axes([0,0,1,1])
#scattering is mostly used in clustering application
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
```

##3D-Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x =np.sin(z)
y =np.cos(z)

ax.scatter(x,y,z,c=z,cmap='Blues')

plt.show()
```

##Legends

```python
plt.plot(x,y)
plt.legend(['testing legend'],loc='lower left')
plt.show()
```
