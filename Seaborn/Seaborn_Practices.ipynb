##Seaborn<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The Python visualization library Seaborn is based on
matplotlib and provides a high-level interface for drawing
attractive statistical graphics.

**Use the Following Convntion**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

The basic steps to creating plots with Seaborn are:<br>
1. Prepare some data<br>
2. Control figure aesthetics<br>
3. Plot with Seaborn<br>
4. Further customize your plot

Note: Seaborn has some built-in datasets <br>
eg-1: total bill vs tip dataset<br>
eg-2: iris dataset<br>
eg-3: titanic dataset<br>
eg-4:house price dataset

**Total bill vs tip dataset**

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head() 
```

using above data we are going to predict how much tip they are going to give

**Visualize the data**

```python
#setting a theme for the plots
sns.set_theme()
```

```python
sns.relplot(data=tips,
           x='total_bill' ,
           y='tip',
           col='time',
           hue='smoker',
           style='smoker',
           size='size')
```

**Iris dataset**

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

using above data we will predict the what species it belongs 

**Scatter Plot**

```python
sns.scatterplot(x='sepal_length',
                y='petal_length',
                hue='species',
                data=iris)
```

```python
sns.scatterplot(x='sepal_length',
                y='petal_width',
                hue='species',
                data=iris)
```

**Titanic dataset**

```python
#loading the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

using above information we will find the person is survived or not 

```python
titanic.shape
```

**Count Plot**

```python
sns.countplot(
    x ='class',
    data=titanic
)
```

```python
#how many survived
sns.countplot(
    x='survived',
    data=titanic
)
```

**Bar Chart**

```python
sns.barplot(
    x='sex',
    y='survived',
    hue='class',
    data=titanic
)
```

**House Price Dataset**

```python
#house price prediction
from sklearn.datasets import load_boston
house_boston =load_boston()

house =pd.DataFrame(house_boston.data,
                    columns=house_boston.feature_names)
house['PRICE'] =house_boston.target
```

```python
house.head()
```

**Distribution Plot**

```python
sns.distplot(house['PRICE'])
```

**Correlation**<br>
1.positive correlation<br>
2.negative correlation


```python
correlation = house.corr()
```

```python
#constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,
            cbar=True,    
            square=True,
            fmt='.1f', # floating point 0.1 will increase
            annot=True, # names in the side
            annot_kws={'size':8},
            cmap='Reds')
```
